```python
import numpy as np
```

```python
from detect_yolo import detect_yolo_from_image
```

```python
from coco_utils import convert_bbox_to_metric, convert_metric_to_box
from coco_utils import metric_map
from coco_utils import load_coco_file
```

```python
annFile='dataset/train/annotations/COCO_json/coco_annotations_train.json'
```

```python
coco=load_coco_file(annFile)
```

```python
resultFile='dataset/submission_example.json'
```

```python
result_coco = load_coco_file(resultFile)
```

```python
folder = 'dataset/train/images/'
```

```python
result = detect_yolo_from_image(folder)
```

```python
#Запишем результаты по изображениям
for i in coco.imgs:
    d_img = coco.imgs[i]
    result_bbox = []
    result_confidence = []
    for j in range(len(result)):
        bbox, confidence, filename = result[j]
        if d_img['file_name'] == filename:
            result_bbox.append(bbox)
            result_confidence.append(confidence)
    coco.imgs[i]['predict_bbox'] = result_bbox
    coco.imgs[i]['predict_confidence'] = result_confidence
```

```python
list_frames = []
for i in coco.imgs:
    d_img = coco.imgs[i]
    d_ann = coco.imgToAnns[i]
    
    list_gt = []
    
    for ann in d_ann:
        s = convert_bbox_to_metric(ann['bbox'])
        x1,y1,x2,y2 = s
        list_gt.append( np.array((x1,y1,x2,y2,0,0,0)) )
        
    list_gt = np.array(list_gt)
        
    list_preds = []
    for j, bbox in enumerate(d_img['predict_bbox']):
        s = convert_bbox_to_metric(bbox)
        x1,y1,x2,y2 = s
        #list_preds.append( np.array((x1,y1,x2,y2,0,d_img['predict_confidence'][j])) )
        list_preds.append( np.array((x1,y1,x2,y2,0,1)) )
        
    list_preds = np.array(list_preds)
        
    list_frames.append((list_preds, list_gt))
```

```python
metric_map(list_frames)
```
